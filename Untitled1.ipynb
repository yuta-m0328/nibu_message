```python
import pandas as pd
import pprint
```

```python
df = pd.read_csv("message_form.csv")[["お名前","メッセージ(50文字以内でお願いします)"]]
```

```python
df
```

```python
all_list = []
for row in df.iterrows():
    # print(row[1]["お名前"])
    name, message = row[1]["お名前"], row[1]["メッセージ(50文字以内でお願いします)"]
    all_list.append({"text": message, "recipient": name})
```

```python
print(str(all_list).replace("'text'","text").replace("'recipient'","recipient").replace("},", "},\n"))
```
